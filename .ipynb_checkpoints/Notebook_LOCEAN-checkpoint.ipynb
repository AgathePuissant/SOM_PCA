```matlab
addpath(genpath('SOM-Toolbox/'));

%Liste des noms des pigments et profondeurs pour l'affichage dans les graphes
noms_pigments = ["total chla", "DVchla", "19hex", "fucox", "perid", "zeax"];
noms_varsat =["chla_sat","RRS412","RRS443","RRS490","RRS555","sst","KD490","KDPAR","PAR","ZEU","ZHL"];
profondeurs = ["5.0","8.35","13.92","23.23","38.75","64.63","107.81","179.85","300.0"];

noms_pigprof = [];
for i=1:length(noms_pigments)
    for j=1:length(profondeurs)
        noms_pigprof = [noms_pigprof, sprintf("%s_%s",[strrep(strrep(noms_pigments(i)," ","_"),"19hex","X19hex"),strrep(profondeurs(j),".","_")])];
    end
end
noms_tout = [noms_pigprof,noms_varsat];
profondeurs = str2double(profondeurs) ;
```

```matlab
T = readtable('data.csv');
dates = readtable('dates.csv');
complete = readtable('completed.csv');
BODC_reel = readtable('data_bodc.csv');

%Noms des stations
stations_incomplete = T(:,ismember(T.Properties.VariableNames,["combined_station"]));
%Longitude et Latitude
longitude = T(:,ismember(T.Properties.VariableNames,["long_sat"]));
latitude = T(:,ismember(T.Properties.VariableNames,["lat_sat"]));

idx = ismember(T.Properties.VariableNames,noms_pigprof);
idx = idx(2:size(complete,2)+1);
T(:,ismember(T.Properties.VariableNames,noms_pigprof)) = complete(:,idx);
T = T(:,ismember(T.Properties.VariableNames,noms_tout));

liste_station_bodc_reel = unique(BODC_reel(:,9));
liste_station_bodc_reel = liste_station_bodc_reel(ismember(liste_station_bodc_reel,stations_incomplete),:);
T_BODC_reel = T(ismember(stations_incomplete,liste_station_bodc_reel),:);
```

```matlab
for pig_ind=1:length(noms_pigments)
    pig=(pig_ind-1)*length(profondeurs)+1:(pig_ind-1)*length(profondeurs)+length(profondeurs);
    data_interm = T{:,pig};
    moy = mean(data_interm);
    var = std(data_interm);
    data_interm = (data_interm-moy)./var;
    center_parameter_interm(:,pig_ind)=moy';
    scale_parameter_interm(:,pig_ind)=var';
    [COEFF, SCORE, LATENT, TSQUARED, EXPLAINED, MU] = pca(data_interm,'Centered','off');
    nbaxes = 2;
    coord(:,(pig_ind-1)*nbaxes+1:(pig_ind-1)*nbaxes+nbaxes) = SCORE(:,1:nbaxes);
    loadings_pig_interm(:,(pig_ind-1)*nbaxes+1:(pig_ind-1)*nbaxes+nbaxes) = COEFF(:,1:nbaxes);
end
```

```matlab
newT=table('Size',[1614,nbaxes*6],'VariableTypes',repmat("double",[1,nbaxes*6]));
newT{:,:}=coord;
T=[newT,T(:,55:end)];
profondeurs=[1:nbaxes];
loadings_pig=table('Size',[9,nbaxes*6],'VariableTypes',repmat("double",[1,nbaxes*6]));
loadings_pig{:,:}=loadings_pig_interm;
scale_parameter=table('Size',[9,6],'VariableTypes',repmat("double",[1,6]));
scale_parameter{:,:}=scale_parameter_interm;
center_parameter=table('Size',[9,6],'VariableTypes',repmat("double",[1,6]));
center_parameter{:,:}=center_parameter_interm;

%stations BODC AMT
BODC = readtable('data_bodc.csv');
liste_station_bodc = unique(BODC(:,9));
liste_station_bodc = liste_station_bodc(ismember(liste_station_bodc,stations_incomplete),:);
T_BODC = T(ismember(stations_incomplete,liste_station_bodc),:);
```

```matlab
data=som_data_struct(T{:,:},'label_names',T.Properties.VariableNames,'comp_names',T.Properties.VariableNames);
%normalisation par la variance
datanorm=som_normalize(data,'var');
%creer la carte initiale
sMap=som_make(datanorm,'munits',400,'tracking',0,'training','Long');
sMap=som_batchtrain(sMap,datanorm,'radius',[100 1],'trainlen',80,'tracking',0);
sMap=som_batchtrain(sMap,datanorm,'radius',[1 0.5],'trainlen',250,'tracking',0);
sMap=som_batchtrain(sMap,datanorm,'radius',[0.5 0.1],'trainlen',200,'tracking',0);
sMap=som_batchtrain(sMap,datanorm,'radius',[0.1 0.001],'trainlen',100,'tracking',0);
```

```matlab
%plot inline
som_show(sMap,'umat',1:size(sMap.codebook,2));%pour visualiser la matrice des distance (UMatrice)
```

```matlab
%plot inline
som_show(sMap,'empty','Hitmap');
hits = som_hits(sMap,datanorm);
som_show_add('hit',hits,'Marker','lattice','MarkerColor','m','Text','on','TextColor','k');
```

```matlab
Ss= som_denormalize(sMap) ;
division=length(profondeurs);
nbpigments = length(noms_pigments);
unite = '';

for i = 0:division:division*nbpigments-1
    figure;
    [ha,~]=tight_subplot(mod(division,3)+fix(division/3),3,[.1 .1],[.1 .1],[.1 .1]);
    title_map = strrep(Ss.comp_names(i+1:i+division),'_','\_');
    for j = 1:division
        axes(ha(j));
        som_cplane(Ss,Ss.codebook(:,i+j));
        title((title_map(j)),'FontSize',8)
        shading flat
        caxis([min(Ss.codebook(:,i+1:i+division),[],'all') max(Ss.codebook(:,i+1:i+division),[],'all')])
        C=colorbar;
        C.Label.String =unite;
        C.FontSize=8;
        colormap(jet)
    end
    
    for vide = division+1:length(ha)
        subplot(ha(vide)); 
        set(gca,'Color','none','XColor','none','YColor','none');
    end
end


figure;
[ha,~]=tight_subplot(mod(size(sMap.codebook,2)-nbpigments*division,3)+fix((size(sMap.codebook,2)-nbpigments*division)/3),3,[.1 .1],[.1 .1],[.1 .1]);
title_map = strrep(Ss.comp_names(nbpigments*division+1:end),'_','\_');
for i=1:size(sMap.codebook,2)-nbpigments*division
    axes(ha(i));
    som_cplane(Ss,Ss.codebook(:,nbpigments*division+i));
    title((title_map(i)),'FontSize',8)
    shading flat

    C=colorbar;
    colormap(jet)
end

for vide = size(sMap.codebook,2)-nbpigments*division+1:length(ha)
    subplot(ha(vide)); 
    set(gca,'Color','none','XColor','none','YColor','none');
end
```
